```python
#1.Square Pattern
n = 5
for i in range(n):
    print("* " * n)
```

```python
# Right Triangle
n = 5
for i in range(1, n+1):
    print("* " * i) 
```

```python
#Reverse Right Triangle
n = 5
for i in range(n, 0, -1):
    print("* " * i)
```

```python
#pyramid pattern
n = 5
for i in range(1, n+1):
    print(" "*(n-i) + "* " * i)
```

```python
#reverse pyramid
n = 5
for i in range(n, 0, -1):
    print(" "*(n-i) + "* " * i)
```

```python
#Diamond pattern
n = 5

# upper
for i in range(1, n+1):
    print(" "*(n-i) + "* " * i)

for i in range(n-1, 0, -1):
    print(" "*(n-i) + "* " * i)
```

```python
#number triangle
n = 5
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

```python
# Increasing number pyramid
n = 5
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

```python
#zero-one triangle
n = 5
for i in range(1, n+1):
    num = i % 2
    for j in range(1, i+1):
        print(num, end=" ")
        num = 1 - num
    print()
```

```python
#pascal's triangle
n = 5
for i in range(n):
    num = 1
    for j in range(i+1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)
    print()
```
