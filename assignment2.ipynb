```python
# E-commerce Discount Calculator
scenario:an online store offere discounts based on the purchase amount:
10%discount for purchases between $100 and $500
20%discount for purchases above $100
no discount for purchases below $100

Task:write a program that takes the purchase amount as input 
and calculates the discount and final amount to be paid.
```

```python
amount = float(input("enter purchase amount : "))
if 100 <= amount <= 500:
    discount = amount * 0.10
elif amount > 500:
    discount = amount * 0.20
else:
    discount = 0
final_amount = amount - discount

print("discount:",discount)
print("final_amount to pay:",final_amount)

```
